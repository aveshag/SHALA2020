## Part 2 - Drawing correlation and heatmap 
---

### Using the correlation matrix, peform the following:

* Find the correlation matrix for the ```movies```.  
* Draw a **heatmap** with the correlation matrix and display the values of correlation coefficients for each pair of features. Make sure the values are properly visible. You might want to adjust the figure size for this. 
* Create a ```mask``` to remove the duplicate values from the **heatmap**. 

````python
# Find out the correlation matrix 
correlation_mat = movies.corr()

# Draw the heatmap 
plt.figure(figsize = (15,13))
sns.heatmap(correlation_mat, vmin = -1, cmap = 'coolwarm_r', annot = True, linewidths = 0.1) 
plt.show()
````

````python
# Creating a mask for removing duplicates 
plt.figure(figsize = (13,12))
mask = np.zeros_like(correlation_mat, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(correlation_mat, vmin = -1, cmap = 'coolwarm', annot = True, mask = mask, linewidths = 0.1)
plt.show()
````

### Using a scatter plot, perform the following:

* Draw a scatter plot between between ```imdb_rating``` and ```audience_score```, and also fit a regression line.
* Change the color of marker to red, size to 200, transparency to 0.3.    



````python
# Scatter plot to fit the regression line 

plt.figure(figsize = (7,7))
sns.regplot('imdb_rating', 'audience_score', color = 'g', data = movies, fit_reg = True)
plt.title('audience_score vs imdb_rating', fontsize = 13)
plt.xlabel('imdb_rating', fontsize = 13)
plt.ylabel('audience_score', fontsize = 13)
plt.show()
````

````python
# Change marker of the scatter plot

plt.figure(figsize = (7,7))
sns.regplot('imdb_rating', 'audience_score', data = movies, color = 'r',
            marker = "*", fit_reg = True, scatter_kws = {"color":"red", "alpha":0.3, "s":200})

plt.title('audience_score vs imdb_rating', fontsize = 13)
plt.xlabel('imdb_rating', fontsize = 13)
plt.ylabel('audience_score', fontsize = 13)
plt.show()
````

### Using a scatter plot, perform the following:

* Load the dataset ```iris```. This has been done for you.
* Draw a scatter plot between between ```petal_length``` and ```petal_width``` and color it according to the ```species```.
* Change the size of marker to 100 and place the legend in ```lower right```. 

````python
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')
 
# Use the 'hue' argument to provide a factor variable

sns.lmplot(x = 'petal_length', y = 'petal_width', data = df, fit_reg = False, markers = ['o','*','x'],  
           hue = 'species', legend = False, scatter_kws = {"alpha":0.8, "s":100})
 
plt.title('petal_width vs petal_length', fontsize = 14)
plt.xlabel('petal_length', fontsize = 14)
plt.ylabel('petal_width', fontsize = 14)

# Move the legend to an empty part of the plot
plt.legend(loc = 'lower right', fontsize = 14)
plt.show()
````

### Using density plots, perform the following:

* Load the dataset ```iris```. This has been done for you.
* Draw the density plots for all the features ```sepal_width```, ```sepal_length```, ```petal_width```, and ```petal_length```. You can required to use subplots for plotting all the density plots. 
* Try drawing a [CDF](https://en.wikipedia.org/wiki/Cumulative_distribution_function) of these features. There is a [blog on medium](https://medium.com/@rishav.jnit/exploratory-data-analysis-eda-on-iris-dataset-using-python-cadd850c1fc6) on how to plot CDF. It will be covered in tutorial. 

````python
# library and dataset 
df = sns.load_dataset('iris')
fig = plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
sns.kdeplot(df['sepal_width'], shade = True, color = "r")
plt.subplot(2,2,2)
sns.kdeplot(df['sepal_length'], shade = True, color = "b")
plt.subplot(2,2,3)
sns.kdeplot(df['petal_width'], shade = True, color = "m")
plt.subplot(2,2,4)
sns.kdeplot(df['petal_length'], shade = True, color = "g")
plt.show()
````

````python
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
sns.kdeplot(df['sepal_width'], shade = True, color = "r", cumulative = True)
plt.subplot(2,2,2)
sns.kdeplot(df['sepal_length'], shade = True, color = "b", cumulative = True)
plt.subplot(2,2,3)
sns.kdeplot(df['petal_width'], shade = True, color = "m", cumulative = True)
plt.subplot(2,2,4)
sns.kdeplot(df['petal_length'], shade = True, color = "g", cumulative = True)
plt.show()
````

````python
plt.figure(figsize = (10,10))
p1 = plt.subplot(2,2,1)
sns.kdeplot(df['sepal_width'], shade = True, color = "r", label = 'PDF')
sns.kdeplot(df['sepal_width'], shade = True, color = "y", cumulative = True, label = 'CDF')
p1.legend(title="sepal_width")

p2 = plt.subplot(2,2,2)
sns.kdeplot(df['sepal_length'], shade = True, color = "b", label = 'PDF')
sns.kdeplot(df['sepal_length'], shade = True, color = "c", cumulative = True, label = 'CDF')
p2.legend(title="sepal_length")

p3 = plt.subplot(2,2,3)
sns.kdeplot(df['petal_width'], shade = True, color = "g", label = 'PDF')
sns.kdeplot(df['petal_width'], shade = True, color = "k", cumulative = True, label = 'CDF')
p3.legend(title="petal_width")

p4 = plt.subplot(2,2,4)
sns.kdeplot(df['petal_length'], shade = True, color = "m", label = 'PDF')
sns.kdeplot(df['petal_length'], shade = True, color = "y", cumulative = True, label = 'CDF')
p4.legend(title="petal_length")

plt.show()
````
